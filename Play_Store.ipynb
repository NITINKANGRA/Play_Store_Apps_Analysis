# **Removing (+) Sign From Installs Column :**

```python
df['Installs'] = df['Installs'].str.replace('+', '').str.replace(',', '')
df['Installs'] = pd.to_numeric(df['Installs'])
```

```python
df
```

# **Taking One Decimal Point In Rating column :**

```python
df['Rating'] = df['Rating'].round(1)
```

# **Data Visualiation :**

# **Q1: - Which is the most used Android version ?**

```python
android_versions = df['Android Ver'].value_counts()
android_versions
```

```python
num_versions = len(android_versions)
colors = plt.cm.Set3(np.linspace(0, 1, num_versions))
plt.figure(figsize=(15, 6))
plt.bar(android_versions.index, android_versions, color=colors)
plt.xlabel('Android Version')
plt.ylabel('Count')
for i, count in enumerate(android_versions):
    plt.text(i, count, str(count), ha='center', va='bottom')
plt.title('Count of Android Versions')
plt.xticks(rotation=90)
plt.show()
```

# **Q2: - Which all categories of mob apps are existing from this dataset ?**

```python
categories = df['Category'].unique()
categories
```

```python
category_counts = df['Category'].value_counts()
category_counts
```

```python
plt.figure(figsize=(15, 6))
category_counts.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Number of Apps in Each Category')
plt.xticks(rotation=90)
plt.show()
```

# **Q3: - Which is the most installed category of mob app from this dataset ?**

```python
category_installs = df.groupby('Category')['Installs'].count()
category_installs
sorted_categories = category_installs.sort_values(ascending=False)
sorted_categories
```

```python
plt.figure(figsize=(15, 6))
sorted_categories.plot(kind='bar', color='orange')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Count of Installs in Each Category')
for i, count in enumerate(sorted_categories):
    plt.text(i, count, str(count), ha='center', va='bottom')
plt.xticks(rotation=90)
plt.show()
```

# **Q6: - Which are the top 10 mob apps based on ratings ?**

```python
# Group the DataFrame by app and calculate the sum of ratings
app_ratings_sum = df.groupby('App')['Rating'].sum()

# Sort the values in descending order and select the top 10 apps
top_10_apps = app_ratings_sum.nlargest(10)
top_10_apps
```

```python
plt.figure(figsize=(10, 6))

# Define the colors
colors = ['skyblue', 'orange', 'green', 'purple', 'yellow', 'pink', 'blue', 'red', 'cyan', 'magenta']

# Plot the bar graph with colors and a bounding box
top_10_apps.plot(kind='bar', color=colors)
plt.box(on=True)

# Add labels and title
plt.xlabel('App')
plt.ylabel('Sum of Ratings')
plt.title('Top 10 Mobile Apps based on Sum of Ratings')

# Display the graph
plt.show()
```
